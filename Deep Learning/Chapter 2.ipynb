# 何謂感知器？
 - 收到多個訊號後，在當做一個訊號輸出。

***


$y = \left\{\begin{matrix}
0, (w_1x_1 + w_2x_2) \leq \theta
\\ 
1, (w_1x_1 + w_2x_2) >  \theta
\end{matrix}\right.$

 - $x_1$ and $x_2$ are input neuron
 - $w_1$ and $w_2$ are $weight$.
 - $y$ is output neuoun.
 - $\theta$ is theshood.
 -  ◯ : neuron or node.

## 2.3.2 導入權重 (weight) 與 偏權值 (bias)

 - 考量後面的部分，調整公式。
 - 將 $\theta$ 改成 $-b$。

$$y = \left\{\begin{matrix}
0, (b+w_1x_1 + w_2x_2) \leq 0
\\ 
1, (b+w_1x_1 + w_2x_2) >  0
\end{matrix}\right.$$


 - 這裡的 $-b$ 當作 $bias$，$w_1$ 與 $w_2$ 當作 $weight$
 - 在感知器當中，將輸入的訊號乘以權重，如果結果 $> 0$ ， 則輸出 1
 - 在感知器當中，將輸入的訊號乘以權重，如果結果超過 $\leq 0$ ， 則輸出 0

### AND GATE

```python
import pandas as pd
import numpy as np

and_pf = pd.DataFrame({
    "x1" : [0,0,1,1],
    "x2" : [0,1,0,1],
    "y"  : [0,0,0,1]
}
)

and_pf
```

```python
# AND GATE function with theshood: theta.
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7 # weight and theshood
    tmp = x1*w1 + x2*w2
    
    if tmp <= theta:
        return(0)
    elif tmp > theta:
        return(1)
    
    else:
        return("input error!")
```

```python
print("AND(0,0): {0}".format(AND(0,0)))
print("AND(1,0): {0}".format(AND(1,0)))
print("AND(0,1): {0}".format(AND(0,1)))
print("AND(1,1): {0}".format(AND(1,1)))
```

## 2.3.3 利用權重與偏權重執行 AND GATE & NAND GATE

```python
# AND GATE function with bias: b.
def AND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    
    tmp = sum(w*x) + b
    
    if   tmp > 0:
        return(1)
    else:
        return(0)
```

```python
print("AND(0,0): {0}".format(AND(0,0)))
print("AND(1,0): {0}".format(AND(1,0)))
print("AND(0,1): {0}".format(AND(0,1)))
print("AND(1,1): {0}".format(AND(1,1)))
```

### NAND Gate (反及閘)

```python
import pandas as pd
import numpy as np

nand = pd.DataFrame({
    "x1" : [0,0,1,1],
    "x2" : [0,1,0,1],
    "y"  : [1,1,1,0]
}
)

nand
```

```python
def NAND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5]) # 只有 weight & bias（theshold） 不同
    b = 0.7  
    
    tmp = sum(w*x) + b
    
    if   tmp > 0:
        return(1)
    else:
        return(0)
```

```python
print("NAND(0,0): {0}".format(NAND(0,0)))
print("NAND(1,0): {0}".format(NAND(1,0)))
print("NAND(0,1): {0}".format(NAND(0,1)))
print("NAND(1,1): {0}".format(NAND(1,1)))
```

### OR Gate (或閘)

```python
import pandas as pd
import numpy as np
OR.df = pd.DataFrame({
    "x1" : [0,0,1,1],
    "x2" : [0,1,0,1],
    "y"  : [0,1,1,1]
}
)

OR.df
```

```python
def OR(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5]) # 只有 weight & bias（theshold） 不同
    b = -0.2  
    
    tmp = sum(w*x) + b
    
    if   tmp > 0:
        return(1)
    else:
        return(0)
```

```python
print("OR(0,0): {0}".format(OR(0,0)))
print("OR(1,0): {0}".format(OR(1,0)))
print("OR(0,1): {0}".format(OR(0,1)))
print("OR(1,1): {0}".format(OR(1,1)))
```

### Q: 如何設計？
### A: 可組合前面原有的 AND, NAND, OR Gate 來設計 XOR Gate

```python
%matplotlib inline
from PIL import Image
Image.open("./textbook_pic/2-9.png")
```

```python
print("組合如下：NAND Gate & OR Gate 的結果當作 input")
print("再用 AND Gate 串接即可得到 XOR 的結果")
%matplotlib inline
from PIL import Image
Image.open("./textbook_pic/2-11.png")
```

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
xor_pd = pd.DataFrame({
    "x1" : [0,0,1,1],         # 0th layer
    "x2" : [0,1,0,1],         # 0th layer
    "s1(NAND)" : [1,1,1,0],   # 1st layer 
    "s2(OR)" : [0,1,1,1],     # 1st layer 
    "y"  : [0,1,1,0]          # 2nd layer
})

xor_pd
```

```python
print("XOR(0,0): {0}".format(XOR(0,0)))
print("XOR(1,0): {0}".format(XOR(1,0)))
print("XOR(0,1): {0}".format(XOR(0,1)))
print("XOR(1,1): {0}".format(XOR(1,1)))
```
