````python
def get_html(): 
    url = "https://car.autohome.com.cn/config/series/770.html?fbclid=IwAR2IuhOlaVNx_g1kCa7MjJoPC2RPBcfH8ZyCCiOzqKtGMzT2UMB0sWOfWAA" 
    headers = {"User-Agent":"Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.186 Mobile Safari/537.36"}
    with requests.get(url=url, headers=headers, timeout = 5) as res: 
        html = res.content.decode("utf-8")
    return html
````

````python
import re
def get_detail(html): 
    
    #處理汽車參數
    #通過正則表達式的 search 方法，匹配數據，然後調用group(0)，即可得到相關的數據
    
    config = re.search("var config = (.*?)};", html, re.S) 
    option = re.search("var option = (.*?)};", html, re.S)  
    
    #處理汽車參數 
    car_info = "" 
    if config and option : 
        car_info = car_info + config.group(0) + option.group(0) 
        print(car_info)
````

````python
htm_1 = get_html()
````

````python
car_info = get_detail(htm_1)
````
