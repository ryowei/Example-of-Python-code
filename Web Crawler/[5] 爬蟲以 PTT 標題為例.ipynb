````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.ptt.cc/bbs/movie/index.html'

# 假裝自己是使用者
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}

res = requests.get(url, headers = headers)
````

````python
# parse the html
soup = BeautifulSoup(res.text, 'html.parser')
````

````python
print(soup)
````

````python
print(soup.title)
print(soup.title.string)
````

## 搜尋文章標題

```
<div class="title">
<a href="/bbs/movie/M.1570453081.A.001.html">[普雷] 返校: 你是在裝睡 還是真的醒不來</a>
```

````python
# 1. 先採用 .find('div' class_ = 'title' )
tmp1 = soup.find('div', class_ = 'title')
print("tmp1: {0}".format(tmp1))
print("-"*60)
print("tmp1.a.string: {0}".format(tmp1.a.string)) # because <a herf>...
print("-"*60)
print("type(tmp1.a.string): {0}".format(type(tmp1.a.string)))
````

````python
soup_subtitle = soup.find_all('div', class_ = 'title')
print(soup_subtitle) # 可以看到一個 list
````

````python
for tit in soup_subtitle:
    if tit.a.string != None :
        print(tit.a.string)
````

````python
soup_subdate= soup.find_all('div', class_ = ['date'])
print(soup_subdate) # 可以看到一個 list
````

````python
for date in soup_subdate:
    if date.string != None :
        print(date.string)
````

````python
def subtitle(soup_subtitle):
    stit = []
    for tit in soup_subtitle:
        if tit.a.string != None :
            subtitle = tit.a.string
            stit.append(subtitle)
    return stit

def ptt_date(soup_subdate):
    sdate = []
    for date in soup_subdate:
        if date.string != None :
            ptt_date = date.string
            sdate.append(ptt_date)
    return sdate
````

````python
import pandas as pd
df_subtitle = subtitle(soup_subtitle)
df_ptt_data = ptt_date(soup_subdate)
````

````python
print("type(df_subtitle): {0}".format(type(df_subtitle)))
print("type(df_ptt_data): {0}".format(type(df_ptt_data)))
````

````python
df = pd.DataFrame({
    'subtitle' : df_subtitle,
    'date' : df_ptt_data
})
````
