## Seaborn

 - Seaborn 是在 Matplot 的基礎上實作更高階的視覺化 API，可以讓畫圖變得更方便、容易。
 
 - 優點：
   - 畫圖指令非常簡單
   - 圖案好看
   
 - 缺點：
   - 雖然大部分的使用上都沒問題，但有些更進階的圖案很難畫或是沒辦法畫（可用matplot互補）
   - 要用 seaborn 之前要先寫輸入 **import seaborn as sns** 載入
   - 之後只要用一行，就可根據**不同類別**加上**不同顏色**以及**標籤**
   
   
   


```python
# load the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Bokeh

 - 互動式視覺化

```python
# Bokeh
from bokeh.plotting import figure, output_file, show

# 準備資料
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# 輸出成靜態 HTML
output_file("lines.html")

# 創建新的標題和軸圖表
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# 繪製直線圖
p.line(x, y, legend="Temp.", line_width=2)

# 呈現結果
show(p)
```

```python
# 使用以下指令，可知道 matplotlib 有哪些 style
plt.style.available
```

```python
# 根據上述所提供畫圖樣板，套入以下的指令
plt.style.use("seaborn-poster")
d1 = np.random.randint(1, 11, 5) 
plt.pie(d1)
plt.show()
```

```python
sns.set(style="white", context="talk")
rs = np.random.RandomState(7)


# 準備 matplotlib 圖表
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 6), sharex=True)

# 產生連續資料
x = np.array(list("ABCDEFGHI"))
y1 = np.arange(1, 10)
sns.barplot(x, y1, palette="BuGn_d", ax=ax1)
ax1.set_ylabel("Sequential")

# 調整成 diverging 資料
y2 = y1 - 5
sns.barplot(x, y2, palette="RdBu_r", ax=ax2)
ax2.set_ylabel("Diverging")

# 隨機資料
y3 = rs.choice(y1, 9, replace=False)
sns.barplot(x, y3, palette="Set3", ax=ax3)
ax3.set_ylabel("Qualitative")

# 秀出圖片
sns.despine(bottom=True)
plt.setp(f.axes, yticks=[])
plt.tight_layout(h_pad=3)
```

### Pandas 與Seaborn 之間有什麼區別呢？

兩者都是使用了 matplotlib 來作圖，但是有非常不同的設計差異

 - 1. 在只需要簡單地作圖時直接用 Pandas，更豐富的圖就可以使用 Seaborn
 - 2.  Pandas 的作圖函數 **沒有太多的參數** 來調整圖形，所以必須要深入了解 matplotlib
 - 3. Seaborn 的作圖函數 **有大量的參數** 來調整圖形，所以不需要太深入了解 matplotlib
 - 4. Seaborn 的 [API](https://stanford.edu/~mwaskom/software/seaborn/api.html) #style-frontend
